Cleaned text data, which is preprocessed by converting all text to lower case,removing punctuation, and removing stopwords using NLTK stopwards list.

After cleaning the text, it is tokenized into words using'RegxpTokenizer', which splits the text into individual words.

```python
# Write your code here
import pandas as pd
imdb_data=pd.read_csv('https://raw.githubusercontent.com/Namratha-Reddy-15/Namratha_Reddy_INFO-5731/main/movie_reviews_with_sentiments%20.csv')
print(imdb_data.shape)


```

```python
import nltk  # Import the NLTK library for natural language processing
from nltk.corpus import stopwords  # Import stopwords for removal of stopwords

# Assuming `imdb_data` is your DataFrame containing the `Clean_Text` column
# Fill missing values in the 'Clean_Text' column with an empty string to avoid errors
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].fillna('')

# Remove punctuation from the 'Clean_Text' column
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].str.replace('[^\w\s]', '', regex=True)

# Convert text to lowercase
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# Download stopwords if not already downloaded
nltk.download('stopwords')

# Define stopwords
stop = stopwords.words('english')

# Remove stopwords from the 'Clean_Text' column
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

# Display the cleaned data
print(imdb_data[['Clean_Text']].head())

```

```python
import pandas as pd
from nltk.tokenize import RegexpTokenizer
from gensim import corpora

# Initialize an empty list to store tokenized words from the documents
words = []

# Create a tokenizer using regular expressions to split text into words
tokenizer = RegexpTokenizer(r'\w+')

# Iterate through the 'Clean_Text' column of imdb_data
for text in imdb_data['Clean_Text']:
    # Tokenize the text using the tokenizer and append the result to the words list
    tokenized_text = tokenizer.tokenize(text)
    words.append(tokenized_text)

# Create a dictionary from the list of tokenized words using Gensim
# This dictionary maps each unique word to a unique integer ID
dictionary = corpora.Dictionary(words)

# Create a corpus from the tokenized words
# The corpus is a list of documents, where each document is represented by a list of (word_id, word_count) tuples
# word_id is the unique integer ID of each word (as per the dictionary), and word_count is the frequency of the word in the document
corpus = [dictionary.doc2bow(word_list) for word_list in words]
```

```python
import gensim  #LDA model
model = gensim.models.ldamodel.LdaModel(corpus, num_topics = 10, id2word = dictionary,passes = 20)
```

```python
print(model.print_topics(num_topics = 10, num_words = 5))
```

```python
import pandas as pd
import nltk

# Convert Clean_Text column to strings and fill NaN values with an empty string
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].astype(str).fillna('')

# Remove punctuation
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].str.replace('[^\w\s]', '', regex=True)

# Convert text to lowercase
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].str.lower()

# Download stopwords and remove them
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
imdb_data['Clean_Text'] = imdb_data['Clean_Text'].apply(lambda x: " ".join(word for word in x.split() if word not in stop))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf= TfidfVectorizer(ngram_range=(1,2), max_features=1000)
tf_idf.fit(imdb_data['Clean_Text'])
x_values =  tf_idf.transform(imdb_data['Clean_Text'])
y_values = imdb_data['Sentiment']


# Split the training data to training and validating data
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x_values, y_values, test_size = 0.2)
```

```python
from sklearn import metrics

def evaluation(y_pred, y_test):
    Accuracy = metrics.accuracy_score(y_pred, y_test)
    Recall = metrics.recall_score(y_pred = y_pred, y_true = y_test, pos_label='positive', average='micro') # micro calculates total true positives, false negatives and false positives
    Precision = metrics.precision_score(y_pred = y_pred, y_true = y_test, pos_label='positive', average='micro') # micro calculates total true positives, false negatives and false positives
    F1 = 2 * (Precision * Recall) / (Precision + Recall) # Formula for F1 Score
    print("Accuracy: ", Accuracy.round(4))
    print("Recall:", Recall.round(4))
    print("Precision:", Precision.round(4))
    print("F-1 score:", F1.round(4))

```

```python
from sklearn import svm, metrics
from sklearn.model_selection import cross_val_score, KFold
from sklearn.impute import SimpleImputer
import numpy as np

# Assuming x_train and y_train contain the training data and labels
# Assuming x_valid and y_valid contain the validation data and labels

# Define imputer for missing values (e.g., replacing with mean values)
imputer = SimpleImputer(strategy='mean')
x_train_imputed = imputer.fit_transform(x_train)
x_valid_imputed = imputer.transform(x_valid)

# Create an SVM classifier
svm_implement = svm.SVC()

# Fit the model on the imputed training data
svm_implement.fit(x_train_imputed, y_train)

# Predict using the validation data
y_pred_valid = svm_implement.predict(x_valid_imputed)

# Define an evaluation function
def evaluation(y_pred, y_test):
    # Calculate accuracy
    accuracy = metrics.accuracy_score(y_test, y_pred)

    # Calculate recall, precision, and F1 score
    recall = metrics.recall_score(y_test, y_pred, average='micro')
    precision = metrics.precision_score(y_test, y_pred, average='micro')
    f1 = metrics.f1_score(y_test, y_pred, average='micro')

    # Print the metrics
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"F1 Score: {f1:.4f}")

# Evaluate the predictions
evaluation(y_pred_valid, y_valid)

# Perform cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=22)
cross_val_scores = cross_val_score(svm_implement, x_valid_imputed, y_valid, cv=kf)

# Print the average cross-validation score
print(f"Cross Validation Score: {cross_val_scores.mean():.4f}")
```

DistilBERT, a pre-trained language model (PLM) from the Hugging Face Repository, has been selected for this challenge. With about 40% fewer parameters, DistilBERT is a speedier and more resource-friendly variant of BERT (Bidirectional Encoder Representations from Transformers). Its performance in different NLP tasks is not greatly affected by this decrease in parameters.
DistilBERT was pre-trained on the same vast dataset as BERT, which included the BookCorpus and Wikipedia, guaranteeing a solid grasp of a wide range of subjects and situations. In general, natural language understanding tasks are where it shines. Without requiring further domain-specific training, it is used directly in a zero-shot environment, making use of its pre-existing pre-trained information to predict sentiment. This method offers a respectable level of accuracy in sentiment analysis in a fast and effective manner.Overall, DistilBERT's balance of performance and efficiency makes it a suitable choice for the task at hand.


```python
import pandas as pd
from transformers import pipeline
from sklearn import metrics
import numpy as np

# Load the CSV file from the link
csv_url = 'https://raw.githubusercontent.com/Namratha-Reddy-15/Namratha_Reddy_INFO-5731/main/movie_reviews_with_sentiments%20.csv'
data = pd.read_csv(csv_url)

# Define columns for text and ground truth labels
text_column = 'Clean_Text'  # Column name containing texts
label_column = 'Sentiment'  # Column name containing ground truth labels

# Extract texts and ground truth labels from the data
x_valid = data[text_column].tolist()
y_valid = data[label_column].tolist()

# Load a smaller model for sentiment analysis to potentially reduce loading time
model_name = "distilbert-base-uncased"
sentiment_pipeline = pipeline("sentiment-analysis", model=model_name)

# Perform sentiment analysis on x_valid using the chosen model
predictions = []
for text in x_valid:
    # Pass the original text to the pipeline and use truncation if necessary
    result = sentiment_pipeline(text, truncation=True, max_length=512)[0]
    predicted_label = result['label']
    predictions.append(predicted_label)

# Convert predictions to a numpy array
predictions = np.array(predictions)

# Function to evaluate the model's performance
def evaluate_model(predictions, y_true):
    accuracy = metrics.accuracy_score(y_true, predictions)
    recall = metrics.recall_score(y_true, predictions, average='micro')
    precision = metrics.precision_score(y_true, predictions, average='micro')
    f1_score = metrics.f1_score(y_true, predictions, average='micro')

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"F1 Score: {f1_score:.4f}")

# Evaluate the model's performance
evaluate_model(predictions, y_valid)
```

Advantages:
Lightweight Model: DistilBERT offers a balance between speed and performance, making it a suitable option for the task.
Zero-shot Capability: The model can be applied directly to sentiment analysis without needing fine-tuning.

Disadvantages:
Limited Task Adaptation: Zero-shot settings may not yield the best performance compared to fine-tuned models.
Potential Bias: The model may inherit biases from its training data, affecting the outcomes.

Challenges Encountered:
Data Preprocessing: Proper formatting and tokenization of data were crucial for successful predictions.
Zero-shot Limitations: The model's zero-shot performance may not be as high as that of models fine-tuned on specific data.
